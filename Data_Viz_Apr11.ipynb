```python
x = np.array([12, 24, 29, 34, 45, 50, 55, 61])
y = np.array([20, 24, 30, 38, 41, 48, 52, 66])
z = np.array([67, 61, 56, 52, 47, 35, 20, 10])
```

```python
# line plot
```

```python
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("y values")

plt.plot(x, y, marker = "o", mfc = "red");
plt.grid();
```

```python
# bar chart
```

```python
course = ["python", "lua", "java", "c#", "c++"]

students = [56, 45, 30, 51, 70]
```

```python
plt.title("student enrollment")
plt.xlabel("course name")
plt.ylabel("no of students")

plt.bar(course, students)
plt.grid()
```

```python
df = pd.read_csv("student.csv")
```

```python
plt.title("student marks")
plt.xlabel("student name")
plt.ylabel("marks")

plt.barh(df['name'], df['marks'])
plt.grid()
```
